```python
import pandas as pd
df = pd.read_csv("Laliga.csv")
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header

```

```python
df.describe()
```

```python

# 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)
df.replace("-", 0, inplace=True)
df = df.fillna(0)
df.head()
```

```python
# 2. Print all the teams which have started playing between 1930-1980. (5 points)
df[start_year_col] = pd.to_numeric(df[start_year_col], errors="coerce")
filtered_df = df[(df["Debut"] >= 1930) & (df["Since/LastApp"] <= 1980)]
filtered_df.head()

```

```python
# 3. Print the list of teams which came Top 5 in terms of points (2.5 points)
points_col = "Points"
df[points_col] = pd.to_numeric(df[points_col], errors="coerce")
sorted_df = df.sort_values(by=points_col, ascending=False)
top_5_teams = sorted_df.head(5)
print(top_5_teams["Team"])

```

```python
# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
def goal_diff_count(df):
    # Convert goals columns to numeric
    df["GoalsFor"] = pd.to_numeric(df["GoalsFor"], errors="coerce")
    df["GoalsAgainst"] = pd.to_numeric(df["GoalsAgainst"], errors="coerce")

    df["GoalDifference"] = df["GoalsFor"] - df["GoalsAgainst"]
    goal_diff_dict = dict(zip(df["Team"], df["GoalDifference"]))

    max_team = df.loc[df["GoalDifference"].idxmax()]["Team"]
    min_team = df.loc[df["GoalDifference"].idxmin()]["Team"]

    return goal_diff_dict, max_team, min_team

goal_diff_dict, max_goal_diff_team, min_goal_diff_team = goal_diff_count(df)

print("Goal Differences:")
for team, diff in goal_diff_dict.items():
    print(f"{team}: {diff}")

print("Team with Maximum Goal Difference:", max_goal_diff_team)
print("Team with Minimum Goal Difference:", min_goal_diff_team)


```

```python
# 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)
df["GamesPlayed"] = pd.to_numeric(df["GamesPlayed"], errors="coerce")
df["GamesWon"] = pd.to_numeric(df["GamesWon"], errors="coerce")

df["WinningPercent"] = (df["GamesWon"] / df["GamesPlayed"]) * 100
df["WinningPercent"].fillna(0, inplace=True)

sorted_df = df.sort_values(by="WinningPercent", ascending=False)
top_5_teams = sorted_df.head(5)
print("Top 5 Teams with Highest Winning Percentages:")
print(top_5_teams[["Team", "WinningPercent"]])

```

```python

# 6. Group teams based on their “Best position” and print the sum of their points for all positions (10 points)
df[points_col] = pd.to_numeric(df["Points"], errors="coerce")
position_points_sum = df.groupby("BestPosition")[points_col].sum()
print("Sum of Points for All Positions:")
print(position_points_sum)
```
